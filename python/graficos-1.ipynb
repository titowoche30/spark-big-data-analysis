```python
import numpy as np
import pandas as pd
import boto3
import io 
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("seaborn-whitegrid")
```

```python
ACCESS_KEY = ''
SECRET_KEY = ''
SESSION_TOKEN = ''
```

```python
client = boto3.client(
    's3',
    aws_access_key_id=ACCESS_KEY,
    aws_secret_access_key=SECRET_KEY,
    aws_session_token=SESSION_TOKEN
)
```

```python
def ler_csv(obj,col1,col2):
    df = pd.read_csv(io.BytesIO(obj['Body'].read()), encoding='utf8')
    df.columns = [col1,col2]
    return df
```

# a)

```python
obj_1_a_manha = client.get_object(Bucket='trab4-bucket',Key='datasets/questoes/1/a_manha.csv/1-a_manha.csv')
df_1_a_manha = ler_csv(obj_1_a_manha,'hashtag','contagem')

obj_1_a_tarde = client.get_object(Bucket='trab4-bucket',Key='datasets/questoes/1/a_tarde.csv/1-a_tarde.csv')
df_1_a_tarde = ler_csv(obj_1_a_tarde,'hashtag','contagem')

obj_1_a_noite = client.get_object(Bucket='trab4-bucket',Key='datasets/questoes/1/a_noite.csv/1-a_noite.csv')
df_1_a_noite = ler_csv(obj_1_a_noite,'hashtag','contagem')
```
